```python
import pandas as pd
```

```python
data_df = pd.read_csv("DOB_Job_Application_Filings.csv", sep = ',', engine = 'python')
```

```python
data_df.head()
```

```python
data_df["Fully Permitted"]
```

```python
start_date = "01/01/2019"
```

```python
end_date = "11/22/2021"
```

```python
after_start_date = data_df["Fully Permitted"] >= start_date
```

```python
before_end_date = data_df["Fully Permitted"] <= end_date
```

```python
between_two_dates = after_start_date & before_end_date
```

```python
filtered_dates = data_df.loc[between_two_dates]
```

```python
print(filtered_dates)
```

```python
filtered_dates.columns
```

```python
proposed_over_seven = filtered_dates['Proposed No. of Stories'] >= 7
```

```python
exisiting_over_seven = filtered_dates['ExistingNo. of Stories'] >= 7
```

```python
over_seven = proposed_over_seven & exisiting_over_seven
```

```python
job_market_19_21 = filtered_dates.loc[over_seven]
```
