```python
import matplotlib.pyplot as plt
```

```python
x = ["2019", "2020", "2021"]
```

```python
jobs_btwn_7_9_19_21 = [133, 141, 156]
jobs_over_10_19_21 = [115, 77, 75]

plt.bar(x, jobs_btwn_7_9_19_21, 0.8, label="btwn 7 & 9 stories high")
plt.bar(x, jobs_over_10_19_21, 0.8, bottom=jobs_btwn_7_9_19_21, label="over 10 stories high")



plt.xlabel("Year")
plt.ylabel("Fully Permitted Jobs")
plt.title("Fully Permitted Jobs between 7 & 9 stories high vs greater than 10 stories high per year:")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
