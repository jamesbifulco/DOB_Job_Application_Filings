# DOB Job Applications Filings Job Market Data Analysis

- This notebook is an exploration of the NYC construction job market in the time period of 2019-2021.
- The data has been provided by the NYC Open Data collection of public databases. 
- Each row of data in this initial file will represent a construction job that has been fully permitted within this time period and is included in the job type category "NB" which denotes that the job is a new building. 
- Throughout the notebook, further filters will be applied to this dataset to analyse the job market to compare jobs that are between 7 and 9 stories and jobs that are 10 stories or above.

```python
import pandas as pd
```

```python
data_df = pd.read_csv("DOB_Job_App_Filings_2019_2021.csv", sep = ',', engine = 'python')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Total Job Market 2019-2021

```python
data_df.head()
```

```python
df = data_df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
#Checking to confirm that each of the jobs in this dataset have been fully permitted
```

```python
df['Fully Permitted'].isnull().values.any()
```

```python
#Each job has been fully permitted. 
```

```python
#Converting the fully permitted column from a string format (text) to a datetime format, to be readable in python.
```

```python
df['Fully Permitted'] = pd.to_datetime(df['Fully Permitted'], errors ="coerce", infer_datetime_format=True)
```

```python
df['Fully Permitted'].iloc[1:50]
```

- Now we will get a record of the total number of fully approved NB jobs there are in this time period regardless of the proposed number of floors:

```python
index = df.index
number_of_rows = len(index)
print(number_of_rows)
```

```python
#The dataset will now be filtered to show how many of these jobs 10 stories high or greater.
```

```python
proposed_over_ten = df['Proposed No. of Stories'] >= 10
```

```python
job_market_19_21_10_plus = df.loc[proposed_over_ten]
```

```python
job_market_19_21_10_plus.head()
```

```python
#Now we will confirm that no jobs are under 10 stories to see the unique values of this column by applying the 
#   .unique() method
```

```python
job_market_19_21_10_plus['Proposed No. of Stories'].unique()
```

- Now we will calculate how many jobs of the 697 are 10 stories or higher:

```python
index = job_market_19_21_10_plus.index
number_of_rows = len(index)
print(number_of_rows)
```

```python
#The dataset will now be filtered to show how many NB jobs fully permitted '19-'21 are between 7 and 9 stories.
```

```python
proposed_over_seven = df['Proposed No. of Stories'] >= 7

proposed_under_nine = df['Proposed No. of Stories'] <= 9

btwn_seven_and_nine = proposed_over_seven & proposed_under_nine

job_market_19_21_btwn_7_and_9 = df.loc[btwn_seven_and_nine]

job_market_19_21_btwn_7_and_9.head()
```

- Now we will calculate how many jobs of the 697 are between 7 and 9 stories high:

```python
index = job_market_19_21_btwn_7_and_9.index
number_of_rows = len(index)
print(number_of_rows)
```

```python
#Now we will confirm that no jobs are out of this range to see the unique values of this column by applying the 
#   .unique() method
```

```python
job_market_19_21_btwn_7_and_9['Proposed No. of Stories'].unique()
```

```python
#To get a better breakdown of these numbers we will filter the data by the individual years in our timeframe
#and repeat the process from above to calculate the increase of market share for each year. 
```

- In 2019-2021, there were a total of 697 fully permitted jobs designated as new buildings ("NB")
- 267 of those jobs were 10 stories or higher
- 430 of those jobs were between 7 and 9 stories high.
