```python
import pandas as pd
```

```python
df = pd.read_csv('job_market_19_21.csv', sep = ',', engine = 'python')
```

```python
nineteen_start_date = "2019-01-01"

nineteen_end_date = "2019-12-31"

after_nineteen_start_date = df["Fully Permitted"] >= nineteen_start_date

before_nineteen_end_date = df["Fully Permitted"] <= nineteen_end_date

nineteen_full_year = after_nineteen_start_date & before_nineteen_end_date

nineteen_job_market = df.loc[nineteen_full_year]

nineteen_job_market.head()
```

```python
index = twenty_job_market.index
number_of_rows = len(index)
print(number_of_rows)
```
